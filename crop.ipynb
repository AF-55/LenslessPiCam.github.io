```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from lensless.utils.image import gamma_correction
import os
```

```python
img = mpimg.imread('images/dataset_simulation/FZA/000032.png')
h, w = img.shape[:2]
cropped = img[:, int((w-h+50)/2):int((w+h-50)/2)]
cropped = gamma_correction(cropped, gamma=1.5)
cropped = (cropped - cropped.min()) / (cropped.max() - cropped.min())
plt.imshow(cropped)
```

```python
for dir in ['tape', 'MLS', 'MURA', 'phase', 'FZA']:
    for img_dir in os.listdir(f'images/dataset_simulation/{dir}'):
        current_dir = f'images/dataset_simulation/{dir}/{img_dir}'
        img = mpimg.imread(current_dir)
        h, w = img.shape[:2]
        cropped = img[:, int((w-h+50)/2):int((w+h-50)/2)]
        cropped = gamma_correction(cropped, gamma=1.5)
        cropped = (cropped - cropped.min()) / (cropped.max() - cropped.min())
        mpimg.imsave(current_dir, cropped)
```
